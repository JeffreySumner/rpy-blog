# Motivation

Diabetes is a rising global health issue, particularly type 2 diabetes. By 2030, it's estimated that around 350 million people will be affected. This project aims to develop a predictive model that assesses an individual's risk of developing diabetes. By harnessing the comprehensive Behavioral Risk Factor Surveillance System (BRFSS) dataset, our team is crafted an interactive dashboard designed to democratize the identification of risk factors and aid in diabetes prevention. Early detection and treatment are critical. In the current technological landscape, there are a host of tools that allow individuals to learn more about their own health such as blood pressure machines; our tool can be used similarly to provide a first pass to motivate individuals to visit their own doctor.
